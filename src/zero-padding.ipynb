```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from numpy import array
from numpy import argmax
from tensorflow.keras.utils import to_categorical
from sklearn.utils import shuffle
    
def get_data():
    file1 = pd.read_csv("../input/audio-spect/Audio_Classification-MLSP/Audio_Classification-MLSP/annotations.csv")
    lstx = []
    lst_y = []
    for m in range(len(file1["fname"])):
        lstx.append(file1["fname"][m])
        lst_y.append(file1["label"][m])
    lst_x = []
    for m in lstx:
        lst_x.append(np.load("../input/audio-spect/Audio_Classification-MLSP/Audio_Classification-MLSP/train/" + m))
    del file1
    return lst_x, lst_y

def reshape_data(lst_x, lst_y):
    dict1 = {}
    lstyy = []
    c = 0
    for m in lst_y:
        if m not in dict1:
            dict1[m] = c
            c = c+1
        lstyy.append(dict1[m])
    N_classes = c+1
    del(dict1)
    data = lstyy
    data = array(data)
    del(lstyy)
    encoded = to_categorical(data)
    del(data)
    lst_y = np.array(encoded)
    del(encoded)
    for b in range(len(lst_x)):
        lst_x[b] = lst_x[b].tolist()
    count = 0
    for m in range(len(lst_x)):
        ct = len(lst_x[m][0][0])
        if ct > count:
            count = ct
    for n in range(len(lst_x)):
        num_zeros = count - len(lst_x[n][0][0])
        zero = [0]*num_zeros
        for m in range(len(lst_x[n][0])):
            lst_x[n][0][m].extend(zero)
    del(zero)
    lst1=[]
    for m in range(len(lst_x)):
        lst1.append(lst_x[m][0])
    lst1 = np.array(lst1)
    del(lst_x)
    return lst1, lst_y
```

```python
lst_x, lst_y = get_data()
```
